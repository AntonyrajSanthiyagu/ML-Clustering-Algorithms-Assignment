```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
from sklearn.cluster import BisectingKMeans
list1 = []
for i in range(1, 11):
    bisect_means =BisectingKMeans(n_clusters=8, init='random', n_init=1, random_state=None, max_iter=300, verbose=0, tol=0.0001, copy_x=True, algorithm='lloyd', bisecting_strategy='biggest_inertia')
    bisect_means.fit(X)
    list1.append(bisect_means.inertia_)
plt.plot(range(1, 11), list1)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import BisectingKMeans
bisect_means =BisectingKMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_bisect_means = bisect_means.fit_predict(X)
y_bisect_means
```

```python
supervised=pd.DataFrame(dataset)
supervised
```

```python
supervised['Cluster_group']=y_bisect_means
supervised
```

```python
supervised.to_csv("Bisecting_K_Means_cluster.csv",index=False)
```

```python
centroids=bisect_means.cluster_centers_
centroids
```
