```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
X = dataset.iloc[:,3:5].values
X
```

```python
from sklearn.cluster import SpectralClustering
y_SC = SpectralClustering(n_clusters=5, eigen_solver= 'arpack', random_state=50, n_init=10, gamma=1.0, affinity='rbf', n_neighbors=10, assign_labels='discretize')
y_Spectral = y_SC.fit_predict(X)
print(y_Spectral)
```

```python
supervised=pd.DataFrame(dataset)
supervised
```

```python
supervised['Cluster_group']=y_Spectral
supervised
```

```python
supervised.to_csv("Spectral_cluster.csv",index=False)
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
